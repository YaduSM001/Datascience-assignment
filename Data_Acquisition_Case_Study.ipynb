***6: Identify any potential outliers in a numerical column of your choice. Explain
your approach.***

```python
column_name = 'minute'
Q1 = df[column_name].quantile(0.25)
Q3 = df[column_name].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df[column_name] < lower_bound) | (df[column_name] > upper_bound)]
outliers
```

The Interquartile Range (IQR) method is used to identify outliers, where any data point below ùëÑ1‚àí1.5√óùêºùëÑùëÖQ1‚àí1.5√óIQR or above ùëÑ3+1.5√óùêºùëÑùëÖQ3+1.5√óIQR is considered a potential outlier, with ùëÑ1Q1 and ùëÑ3Q3 representing the 25th and 75th percentiles
