# ***Diabetes Dataset EDA:***

### **Data Load**

````python
import pandas as pd

dbt= pd.read_csv('/content/diabetes_model.csv')

print(dbt.head())
````

### ***Univariate Analysis:***
**Analyzation of individual variables**

````python
dbt.hist(figsize=(10,10))
````

In the above histogram the y-axis represent the count of the patient and the x-axis represent the range of the variables.


### ***Bivariate Analysis:***
**Exploration of relationships between pairs of variables**

````python
# Scatter plots
plt.figure(figsize=(10, 6))
plt.scatter(dbt['Pregnancies'], dbt['Glucose'],color="red")
plt.xlabel('Pregnancies')
plt.ylabel('Glucose')
plt.title('Pregnancies vs Glucose')
plt.show()

# correlation matrices
corr_matrix = dbt.corr()
plt.figure(figsize=(10, 8))
import seaborn as sns
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
````

Scatterplot shows the relationship between the pregnencies and glucose and correlation matrix shows the relation between each variables.

### ***Multivariate Analysis:***
**Investigations of interactions among multiple variables**

````python
import seaborn as sns
sns.pairplot(dbt[['Pregnancies', 'Glucose', 'BloodPressure','Outcome']],hue='Outcome',palette='deep',markers='*')

plt.show()
````

In pairplot the outcome shows if the patient has diabetes or not,the x-axis shows the number of patients & y-axis shows the range of variables.
