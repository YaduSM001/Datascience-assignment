```python
def detect_outliers_iqr(data):
    """
    Detect outliers using the Interquartile Range (IQR) method.

    Args:
      data: A pandas Series or DataFrame.

    Returns:
      A list of outlier values.
    """
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    return outliers
for col in hp.select_dtypes(include='number').columns:
  outliers = detect_outliers_iqr(hp[col])
  print(f"Outliers in {col}: {outliers}")
```

```python

def cap_outliers(data, col, lower_bound, upper_bound):

  data[col] = data[col].clip(lower=lower_bound, upper=upper_bound)
for col in hp.select_dtypes(include='number').columns:
  outliers = detect_outliers_iqr(hp[col])
  if outliers.any():
    Q1 = hp[col].quantile(0.25)
    Q3 = hp[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    cap_outliers(hp, col, lower_bound, upper_bound)
```
