```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
print(iris.head())
```

***Create atleast 10 visuals from the data.***

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='species', y='sepal_length', data=iris)
plt.title('Violin Plot of Sepal Length by Species')
plt.show()
```

```python
# 2. Swarm Plot
plt.figure(figsize=(20, 25))
sns.swarmplot(x='species', y='sepal_width', data=iris)
plt.title('Swarm Plot of Sepal Width by Species')
plt.show()
```

```python
# 3. Strip Plot
sns.stripplot(x='species', y='petal_length', data=iris)
plt.title('Strip Plot of Petal Length by Species')
plt.show()
```

```python
# 4. Box Plot
sns.boxplot(x='species', y='petal_width', data=iris)
plt.title('Box Plot of Petal Width by Species')
plt.show()
```

```python
# 5. Bar Plot

sns.barplot(x='species', y='sepal_length', data=iris, estimator=np.mean)
plt.title('Bar Plot of Average Sepal Length by Species')
plt.show()
```

```python
# 6. Count Plot
sns.countplot(x='species', data=iris)
plt.title('Count Plot of Species')
plt.show()
```

```python
# 7. Heatmap
numerical_iris = iris.select_dtypes(include=['float', 'int'])
sns.heatmap(numerical_iris.corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap of Feature Correlations')
plt.show()
```

```python

# 8. Scatter Plot
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
plt.title('Scatter Plot of Sepal Length vs Sepal Width by Species')
plt.show()
```

```python
# 9. Line Chart
sns.lineplot(data=iris.drop(columns='species'))
plt.title('Line Chart of All Features')
plt.show()
```

```python
# 10. Pair Plot
sns.pairplot(iris, hue='species')
plt.suptitle('Pair Plot of All Features', y=1.02)
plt.show()
```
